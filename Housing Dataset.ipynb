```python
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.impute import SimpleImputer
```

## Creating the complex model (f)

### Loading the data

```python
data = fetch_openml(name="house_prices", as_frame=True)

df= data.frame
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

### One-hot encoding the data

```python
df_encoded=pd.get_dummies(df, drop_first=True)

df_encoded.head()
```

Note here that we do not use PCA to reduce the number of features since the goal is explainability and not creating a robust model. (Perhaps this could be mentioned as one of the limitations of LIME?)

### Handling null values via imputation

```python
df_encoded.isna().sum()[df_encoded.isna().sum() > 0]
```

Since the columns having null values are all numerical, imputing using the mean is a good strategy.

```python
imputer = SimpleImputer(strategy="mean")  

df_cleaned = pd.DataFrame(imputer.fit_transform(df_encoded), columns=df_encoded.columns)
```

### Training a black box model

```python
features=list(df_cleaned.columns)

y=df_encoded["SalePrice"]

features.remove("SalePrice")

X=df_cleaned[features]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
black_box_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
black_box_model.fit(X_train, y_train)

print(f"Model Accuracy: {black_box_model.score(X_test, y_test):.2f}")
```
